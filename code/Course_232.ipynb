```python
import requests
from bs4 import BeautifulSoup
rs = requests.session()
res = rs.get('https://www.macromicro.me/macro')
soup = BeautifulSoup(res.text, 'lxml')
token = soup.select_one('.sosume p[data-stk]').get('data-stk')
```
